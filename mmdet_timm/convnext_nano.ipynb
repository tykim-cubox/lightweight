```python
%%capture captured

import mmcv
config = '/home/aiteam/tykim/scratch/lightweight/mmdet_timm/edgenext_image1k_faster_rcnn.py'
cfg = mmcv.Config.fromfile(config)
```

```python
cfg.model.backbone.model_name = 'convnext_nano'
cfg.model.backbone.out_indices=(0, 1, 2, 3)

cfg.model.neck.in_channels=[80, 160, 320, 640]

from mmdet.apis import set_random_seed

cfg.data_root = '/home/aiteam/tykim/scratch/lightweight/mmcv_phone/hold_smartphone'
# timm이 다운받는 위치
cfg.load_from = '/home/aiteam/.cache/torch/hub/checkpoints/convnext_nano_d1h-7eb4bdea.pth'
cfg.work_dir = '/home/aiteam/tykim/scratch/lightweight/mmdet_timm/convext_nano_exps'

cfg.data.samples_per_gpu=8
cfg.data.workers_per_gpu=4

cfg.optimizer.lr = 0.02 * 1
cfg.lr_config.warmup = None
cfg.log_config.interval = 10

# We can set the evaluation interval to reduce the evaluation times
cfg.evaluation.interval = 10
# We can set the checkpoint saving interval to reduce the storage cost
cfg.checkpoint_config.interval = 10

cfg.runner.max_epochs = 300

cfg.seed = 0
set_random_seed(0, deterministic=False)

print(f'Config:\n{cfg.pretty_text}')
```

```python
# 데이터셋과 detector를 init한다.
from mmdet.datasets import build_dataset
from mmdet.models import build_detector
from mmdet.apis import train_detector
```

```python
datasets = [build_dataset(cfg.data.train.dataset)]
```

```python
model = build_detector(cfg.model)
```

```python
model.CLASSES = datasets[0].CLASSES
```

```python
train_detector(model, datasets, cfg, distributed=False, validate=True)
```
